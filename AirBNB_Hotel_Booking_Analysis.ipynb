```python
#import required libraries
import pandas as pd #missing values
import numpy as np #numerical values
import matplotlib.pyplot as plt #graphical
import seaborn as sns #more attractive plots
import warnings #ignore the warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('/content/Attrition_Analytics.csv')
```

```python
df.head()
```

```python
df.shape #columns and rows of data
```

```python
df.describe() #statiscics of data(count, min, max)
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Fill missing values in YearsWithCurrManager
df["YearsWithCurrManager"].fillna(df["YearsWithCurrManager"].median(), inplace=True)
```

```python
sns.set(style="whitegrid", palette="muted")

# --- Attrition Count ---
plt.figure(figsize=(6,4))
sns.countplot(x="Attrition", data=df, palette="Set2")
plt.title("Employee Attrition Count")
plt.xlabel("Attrition Status")
plt.ylabel("Number of Employees")
plt.show()
```

```python
# --- Attrition by Department ---
plt.figure(figsize=(8,5))
sns.countplot(x="Department", hue="Attrition", data=df, palette="coolwarm")
plt.title("Attrition by Department")
plt.xticks(rotation=30)
plt.show()
```

```python
# --- Salary Distribution ---
plt.figure(figsize=(8,5))
sns.histplot(df["MonthlyIncome"], bins=40, kde=True, color="orange")
plt.title("Distribution of Monthly Income")
plt.xlabel("Monthly Income")
plt.ylabel("Count")
plt.show()
```

```python
# --- Average Income by Job Role ---
plt.figure(figsize=(10,5))
income_role = df.groupby("JobRole")["MonthlyIncome"].mean().sort_values(ascending=False)
sns.barplot(x=income_role.index, y=income_role.values, palette="magma")
plt.title("Average Monthly Income by Job Role")
plt.xticks(rotation=45)
plt.ylabel("Average Monthly Income")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Age", y="MonthlyIncome", hue="Attrition", data=df)
plt.title("Age vs Monthly Income (Attrition Highlighted)")
plt.xlabel("Age")
plt.ylabel("Monthly Income")
plt.show()
```

```python
# --- Correlation Heatmap ---
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm", annot=True, fmt=".2f")
plt.title("Correlation Heatmap (Attrition Analytics)")
plt.show()
```
